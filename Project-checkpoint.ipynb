```python
# Helper functions

from sklearn import metrics
# helper function to print accuracy, precision, recall, F1
def print_4_metrics(target_test, predicted):
  print("%-12s %f" % ('Accuracy:', metrics.accuracy_score(target_test,predicted)))
  print("%-12s %f" % ('Precision:', metrics.precision_score(target_test, predicted,labels=None, pos_label=1, average='binary', sample_weight=None)))
  print("%-12s %f" % ('Recall:', metrics.recall_score(target_test, predicted,labels=None, pos_label=1, average='binary', sample_weight=None)))
  print("%-12s %f" % ('F1 Score:', metrics.f1_score(target_test, predicted,labels=None, pos_label=1, average='binary', sample_weight=None)))

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools

# Helper function that allows you to draw nicely formatted confusion matrices
def draw_confusion_matrix(y, yhat, classes):
    '''
        Draws a confusion matrix for the given target and predictions
        Adapted from scikit-learn and discussion example.
    '''
    plt.cla()
    plt.clf()
    matrix = confusion_matrix(y, yhat)
    plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.Reds)
    plt.title("Confusion Matrix")
    plt.colorbar()
    num_classes = len(classes)
    plt.xticks(np.arange(num_classes), classes, rotation=90)
    plt.yticks(np.arange(num_classes), classes)

    fmt = 'd'
    thresh = matrix.max() / 2.
    for i, j in itertools.product(range(matrix.shape[0]), range(matrix.shape[1])):
        plt.text(j, i, format(matrix[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if matrix[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()
```

```python
def lr_schedule(epoch):
    return 0.001 * 0.9**epoch  # Adjust the learning rate as needed

model =Sequential([  
   Embedding(vocabulary_size, 50, input_length=input_length ,embeddings_regularizer="l2"),
     Dropout(0.5),
    BatchNormalization(),
    GlobalAveragePooling1D(),
    Dense(128, activation="relu" , kernel_regularizer='l2'), 
    Dropout(0.5),
    BatchNormalization(),
    Dense(64, activation="relu" , kernel_regularizer='l2'), 
    Dropout(0.5),
    BatchNormalization(),
    Dense(32, activation="relu", kernel_regularizer='l2' ), 
    Dropout(0.4),
    BatchNormalization(),
    Dense(16, activation="relu" , kernel_regularizer='l2' ),
    Dropout(0.3),
    BatchNormalization(),
    Dense(2, activation="softmax"),

])


learning_rate_scheduler = LearningRateScheduler(lr_schedule)

early_stopping = EarlyStopping(monitor="val_loss", patience= 10, restore_best_weights=True, )
model.compile(loss='binary_crossentropy',
              optimizer=Adam(learning_rate=0.001),
              metrics=['accuracy'] )

his  = model.fit(X_train, Y_train,
          epochs=65,
         validation_data=(X_test, Y_test),
                 callbacks= [early_stopping, learning_rate_scheduler],
          batch_size=64, 
                 verbose=True)
```

```python
# tokenizing and padding the data for RNN
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences

max_features = 2000

tokenizer = Tokenizer(max_features, split = ' ')
tokenizer.fit_on_texts(data_clean)
X_tokenized = tokenizer.texts_to_sequences(data_clean)
X_padded = pad_sequences(X_tokenized)

X_train_rnn, X_test_rnn, y_train_rnn, y_test_rnn = train_test_split(X_padded, data_labels, test_size=0.2, random_state=4)
```

```python
X_padded.shape
```

```python
print(X_train_rnn.shape)

print(X_test_rnn.shape)

print(y_train_rnn.shape)
print(y_test_rnn.shape)
```

```python
def lr_schedule(epoch):
    if epoch < 10:
        return 0.0001
    else:
        return 0.001 * 0.9**epoch



model2 = Sequential()

# hyperparams

embed_vec_size = 100 # hyperparam that controls size of embedding vector
num_layers = 200 # also hyperparam
batch_size = 64
learning_rate=0.0001
dropout_rate = 0.3
reg_strength = 0.005
patience = 10

# optimizer = ['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam']
# batch_size = [16,32,64,128,256]

model2.add(Embedding(input_dim=max_features, output_dim=embed_vec_size))
model2.add(BatchNormalization())
model2.add(Dropout(dropout_rate))
model2.add(LSTM(num_layers, dropout=0.3, return_sequences=True))
model2.add(BatchNormalization())
model2.add(LSTM(num_layers-50,dropout=0.3 ,return_sequences=True))  # Additional LSTM layer
model2.add(BatchNormalization())
model2.add(LSTM(num_layers-100 ))  # Additional LSTM layer
model2.add(Dense(units=64, activation='relu', kernel_regularizer="l2"))
model2.add(Dense(units=32, activation='relu'))
model2.add(BatchNormalization())
model2.add(Dropout(0.3))
model2.add(Dense(units=16, activation='relu'))
model2.add(Dense(units=1, activation='sigmoid'))

learning_rate_scheduler = LearningRateScheduler(lr_schedule)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=patience, restore_best_weights=True)

model2.compile(optimizer=Adam(learning_rate=learning_rate), loss='binary_crossentropy', metrics=['accuracy'])
hist = model2.fit(X_train_rnn, y_train_rnn,
                  epochs=100,
                  batch_size=batch_size,
                  validation_data=(X_test_rnn, y_test_rnn)
                  , callbacks=[early_stopping ,learning_rate_scheduler]) 
```

```python
score = model2.evaluate(X_test_rnn, y_test_rnn ,verbose=1) 

print('Test loss: ', score[0])
print('Test accuracy: ', score[1])
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.axhline(score[1] , color = "green")
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation' , "test"], loc='upper left')
stop_epoch = early_stopping.stopped_epoch-patience
plt.axvline(stop_epoch, color='r')

plt.show()
```
